```python
from PIL import Image
import pytesseract
import cv2
import time
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'
```

```python
video = cv2.VideoCapture('/content/segment_3.ts')

if (video.isOpened() == False):
  print("Error while reading Video")
else:
  fps = video.get(cv2.CAP_PROP_FPS)

  frame_ct = video.get(cv2.CAP_PROP_FRAME_COUNT)

while(video.isOpened()):
  ret, frame = video.read()
  if ret == True:
    cv2_imshow(frame)
    key = cv2.waitKey(20)
    if key == ord('q'):
      break
  else:
      break
  video.release()
  cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture('/content/segment_3.ts')  # Replace 'your_video_file.mp4' with your video file path or URL

#target_frame_rate = 9
#frame_delay = int(1000/target_frame_rate)
# Loop through the frames, read, and display them

while True:
    #start_time = time.time()
    ret, frame = cap.read()
    if not ret:
        break
    #cv2_imshow(frame)
    frame = frame[580:650, 690:850]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)


    config = ('-l eng --oem 1 --psm 6')
    text = pytesseract.image_to_string(frame, config=config)
    text = text.split("\n")
    cv2_imshow(gray)
    try:
      print(float(text[0]),float(text[1]))
    except:
      continue

    time.sleep(0.6)
    #end_time = time.time()
    #print(end_time - start_time)

# Release the video capture object
cap.release()
```
